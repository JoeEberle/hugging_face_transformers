## Optional Step 0 - Intitiate Configuration Settings and name the overall solution

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis") 

happy_text = " I can't believe how happy I am this is the most wonderful thing that's ever happened"
sentiment_results = classifier(happy_text)
print(f"text: {happy_text} \nresult: {sentiment_results} ")

sad_text = " I hate this this is the worst thing that's ever happened to me I can't believe how awful this is"
sentiment_results = classifier(sad_text)
print(f"text: {sad_text} \nresult: {sentiment_results} ")

sad_text = " this is not good, but I can make do"
sentiment_results = classifier(sad_text)
print(f"text: {sad_text} \nresult: {sentiment_results} ")

ambivilant_text = "My meal was just so so "
sentiment_results = classifier(ambivilant_text)
print(f"text: {ambivilant_text} \nresult: {sentiment_results} ")

sad_text = "My frog died .... so Yeah "
sentiment_results = classifier(sad_text)
print(f"text: {sad_text} \nresult: {sentiment_results} ")

neutral_text = "an log is on the ground"
sentiment_results = classifier(neutral_text)
print(f"text: {neutral_text} \nresult: {sentiment_results} ")
```
